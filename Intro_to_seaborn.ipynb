**A Scatter / Count plot with hue**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# Specifying hue colors
hue_colors = {"Yes": "black", "No": "red"}

# Scatter Plot
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker", palette=hue_colors)

# Count plot
sns.countplot(x="smoker", data=tips, hue="sex")

plt.show()
```

# **Visualizing Two Quantitive Variables**

- relplot() refers to relational plots
- relplot() create subplots in a single figure


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# We can also add row with col. We can change the order using row_order/col_order
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="smoker")

plt.show()
```

**Line Plots**

- Scatter plot gives information about independent observation
- Line plot represent the same "thing" typically tracked over time

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

air_df_mean = pd.DataFrame({"hour": [1,1,1,1,2], "NO_2_mean":[10.00,11.041667, 21.6667, 10.41667, 33.95833], "location":["East", "North", "South", "West", "East"]})

sns.relplot(x='hour', y="NO_2_mean", data=air_df_mean, kind="line", style="location", hue="location")


plt.show()
```

**Box Plot**
- Shows the distribution of quantitive data
- See median, spread, skewness, and outliers
- Facilitates comparisons between groups

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')

# We can change the order of categorical variable using order parameter
# We can omit the outliers using sym parameter.
'''
Changing the whisker using whis parameter
- By default, the whisker extend to 1.5 * the interquartile range
- Make them extend to 2.0 * IQR: whis=2.0
- Show the 5th and 95th percentiles: whis=[5, 95]
- Show min and max values: whis=[0, 100]
'''
g = sns.catplot(x="time", y="total_bill", data=tips, kind="box", whis=[0, 100])

plt.show()
```

# **Customizing Seaborn Plot**

**FacetGrid vs AxesSubplot objects**

- FacetGrid -> relplot(), catplot() -> Can create subplots
- AxesSubplot -> scallterplot(), countplot(), etc -> Only creates a single plot


```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')

g = sns.catplot(x="time", y="total_bill", data=tips, kind="box", whis=[0, 100])

# Add title in FacetGrid object
g.fig.suptitle("New Title", y=1.03)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')

g = sns.boxplot(x="time", y="total_bill", data=tips)

# Add title in AxesSubplot object
g.set_title("New Title", y=1.03)

# This work with both objects
g.set(xlabel="New X Label", ylabel="New Y label")

# Roate x ticks
plt.xticks(rotation=90)

plt.show()
```
